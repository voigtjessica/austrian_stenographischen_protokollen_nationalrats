```python
# Bibliotecas

import json
import pandas as pd
import requests
import re

#pra validacao
import random

#Abrindo o arquivo que cont√©m os links json com as iformacoes das Sitzungen

with open("dic_links_partlament_at_2019_2023.json", "r") as arquivo:
    sitzungen = json.load(arquivo)
```

```python
#sitzungen

link_value = 'http://www.parlament.gv.at/gegenstand/XXVII/NRSITZ/183?json=true'

#nao lembro o que eu tava fazendo aqui hahaha
link_positions = [i for i, x in enumerate(sitzungen['link']) if x == link_value]

print(link_positions)
print(sitzungen['file'][21])
print(sitzungen['file'][22])
print(sitzungen['file'][23])
```

```python
# Final


#### Trabalhando com todas as urls #####

df_speakers = pd.DataFrame({'title': [], 'url':[], 'topic': [], 'name_politician': [], 'tag_politician': [], 'url_speech': []})
links = sitzungen['link']


for l in links:                                   #Parte 1: request da p√°gina - loop dentro de cada p√°gina
    url = l
    response = requests.get(url)
    json_data = response.json()

    print(url)

    df_sessao = pd.DataFrame({'topic': [], 'name_politician': [], 'url_speech': []})   # cria um DF tempor√°rio para essa secao
    qtde_stages = len(json_data['content'][2]['progress'])                             # v√™ quantos stages tem 

    for i in range(0,qtde_stages):                                                     #loop dentro de cada stage 
        p = json_data['content'][2]['progress'][i]
        if 'speeches' in p:                                                            #Se existem discursos associados a esse stage
            topic = json_data['content'][2]['progress'][i]['text']

            #quantidade de discursos dentro de um tema:
            qtde_discursos = len(json_data['content'][2]['progress'][i]['speeches'])

            for row in range(0, qtde_discursos):                                        #pegando os speeches dentro de um discurso:

                name_politician = json_data['content'][2]['progress'][i]['speeches'][row]['texta']
                url_speech = json_data['content'][2]['progress'][i]['speeches'][row]['protocol']['data']['links'][0]['documents'][0]['link'] 

                df_temp = pd.DataFrame({'topic': [topic], 
                                  'name_politician': [name_politician],
                                  'url_speech': [url_speech]})
                
                df_sessao = pd.concat([df_sessao, df_temp])
        else:
            next
        
    df_sessao['title'] = json_data['meta']['description']
    df_sessao['url'] = url

    df_speakers = pd.concat([df_speakers, df_sessao ])

df_speakers = df_speakers.reset_index()
```
