# Web scraping from austrian Parlament

Nesse primeiro notebook eu estou requisitando os protocolos (Sitzung) de 2019 até 2023 e separando os links nos quais eu fou fazer a requisicao dos json files

```python
# Bibliotecas

import json
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup

# with open("protokole_2000_to_2023.json", "r") as arquivo:
#     protocols = json.load(arquivo)
```

```python
# Importando a busca (eu consegui esse json no site da API do parlament)

url = 'https://www.parlament.gv.at/Filter/api/json/post?jsMode=EVAL&FBEZ=WFP_007&listeId=11070&showAll=true&export=true'

body = {
  "MODUS": ["PLENAR"],
  "NRBRBV": ["NR"],
  "GP": ["XXVII"],
  "R_SISTEI": ["SI"]
}

response = requests.post(url, json=body)

df1 = response.json()

dados = df1['rows']
df = pd.DataFrame(dados, columns=['data', 'sitzung', 'tagesordnung', 'GP', 'R_SISTEI', 'TOP', 'anfrage', 'datum', 'URL', 'art', 'session_overview'])
df
```

```python
# Segunda parte: separando os links das sitzungen

dict_links = {"link": [], "file": []}

## inicio fo loop  
for i in range(0, len(df)):

    # Section overview
    so = df.iloc[i, 10]
    link = 'http://www.parlament.gv.at' + so + '?json=true'


    #criando o nome do arquivo : ano, mes, dia numero da sessao

    data = df.iloc[i,0]
    data = re.sub(r'\.', '', data)
    dia_str = data[:2]
    mes_str = data[2:4]
    ano_str = data[4:]

    # section overview
    so = re.sub(r'^(.*NRSITZ)', '', so)
    so = re.sub(r'\/', '', so)
    so = re.sub(r'\?.*', '', so)

    file = ano_str + '-' + mes_str + '-'  + dia_str + "_" + so

    #adicionando as chaves no meu dicionário:

    dict_links["link"].append(link)
    dict_links["file"].append(file)
```

```python
# Salva o dicionário no arquivo
with open('dic_links_partlament_at_2019_2023.json', 'w') as handle:
    json.dump(dict_links, handle)
```
